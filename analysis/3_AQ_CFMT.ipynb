# Load libraries 

```python
import pandas as pd
import numpy as np
import os
import re


# figures
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes = True)

%matplotlib inline
```

# Paths 

```python
analysispath = os.getcwd()
datapath = analysispath + '/../data/'
```

# Load data 

```python
data = pd.read_csv(datapath + 'holistic_face_abilities.csv')
subject_holistic_index_demographics = pd.read_csv('2_processed_holistic_index_demographics.csv', index_col=0)
```

```python
subject_holistic_index_demographics.head()
```

```python
data = data.drop([0, 1])
data.reset_index(inplace=True, drop=True)
data
```

# Column names 

SC1 is AQ - SC2 is CFMT

```python
AQ_column = 'SC1'
CFMT_column = 'SC2'
```

```python
abilities_cols = [AQ_column, CFMT_column, 'mTurkCode']
abilities_data = data[abilities_cols].reset_index(drop=True)
```

# Assign abilities to subjects 

```python
for subject in subject_holistic_index_demographics.SUBJECTID.unique():
    
    # get mturkcode for that subject
    mturkcode = subject_holistic_index_demographics.loc[subject_holistic_index_demographics['SUBJECTID'] == subject, 'MTURKCODE'].reset_index(drop=True)[0]
    
    # get the original data for that subject
    subject_original_data = abilities_data[abilities_data['mTurkCode'] == str(mturkcode)]
    
    subject_holistic_index_demographics.loc[subject_holistic_index_demographics['SUBJECTID'] == subject, 'AQ'] = subject_original_data[AQ_column]
    subject_holistic_index_demographics.loc[subject_holistic_index_demographics['SUBJECTID'] == subject,'CFMT'] = subject_original_data[CFMT_column]
```

# Save data 

This data now have holistic index, demographics, AQ and CFMT

```python
subject_holistic_index_demographics.to_csv('3_processed_all.csv')
```
