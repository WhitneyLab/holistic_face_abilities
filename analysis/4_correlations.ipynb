# Load libraries 

```python
import pandas as pd
import numpy as np
import os
import re
from scipy import stats 


# figures
import matplotlib.pyplot as plt
import seaborn as sns
import plotFuncs
sns.set(color_codes = True)

%matplotlib inline
```

# Paths 

```python
analysispath = os.getcwd()
datapath = analysispath + '/../data/'
```

# Load data 

```python
data_all = pd.read_csv('3_processed_all.csv')
```

# Correlations 

```python
sns.regplot(data_all['HOLISTIC_INDEX'], data_all['AQ'])
rvalues = stats.pearsonr(data_all['HOLISTIC_INDEX'], data_all['AQ'])
plotFuncs.correlationPlot(plt,'Holistic Index', 'AQ', ' ', rvalues)
```

```python
sns.regplot(data_all['HOLISTIC_INDEX'], data_all['CFMT'])
rvalues = stats.pearsonr(data_all['HOLISTIC_INDEX'], data_all['CFMT'])
plotFuncs.correlationPlot(plt,'Holistic Index', 'CFMT', ' ', rvalues)
```

# Regression 

```python
## Without a constant

import statsmodels.api as sm

# X = df["RM"]
# y = target["MEDV"]

# Note the difference in argument order
model = sm.OLS(data_all['HOLISTIC_INDEX'], data_all['AQ']).fit()
predictions = model.predict(data_all['AQ']) # make the predictions by the model

# Print out the statistics
model.summary()
```
